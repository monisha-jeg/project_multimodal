```python
import pickle
import os, json, pickle

train_dir = "../../../dataset/v1/train/"


state_count_map = {}
state_subtypes = {}
count = 0
for json_name in os.listdir(train_dir)[:]:
	dialogue = json.load(open(train_dir + json_name))
	for utterance_ in dialogue:
		utterance = utterance_["utterance"]
        
		state = utterance_["type"]
		#if state == "answer":
		#	continue
		substate = None
		if state == "question":
			if "question-type" in utterance_:
				state = utterance_["question-type"]
			if "question-subtype" in utterance_:
				substate = utterance_["question-subtype"]

		if state not in state_count_map:
			state_count_map[state] = 1
			if substate != None:
				state_subtypes[state] = set([substate])
		else:
			state_count_map[state] += 1
			if substate != None:
				state_subtypes[state].add(substate)
                
		if count % 1000000 == 1:
			print(count)
		count += 1
            
p#rint(count, state_count_map)
for state, scount in state_count_map.items():
    print "\n", state, scount*100/float(count)
    print state_subtypes.get(state)
```
